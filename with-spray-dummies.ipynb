```python
has_sprayed_dict = {} #empty dict   will fill dict with 10 dummy lists for each spray, has it been sprayed yet 
                         #for the calendar year
                      #then will convert lists to numpy arrays, add them, use as a single 'has_sprayed dummy'

for rowB in range(spray_centers.shape[0]):
    has_sprayed_dict[f'{spray_centers.iloc[rowB, 0]}'] = []      #one list for each spray
    
    for rowA in range(wn_train.shape[0]):
        if (wn_train.iloc[rowA, 0] >= spray_centers.iloc[rowB, 0] and #if the date is at least the spray date
        wn_train.iloc[rowA, 0] <= f'{spray_centers.iloc[rowB, 0][:4]}-12-31' and # and at most the end of the year
        wn_train.iloc[rowA, 2] >= spray_centers.iloc[rowB, 3] and #and latitude is at least min lat
        wn_train.iloc[rowA, 2] <= spray_centers.iloc[rowB, 4] and  #and latitude is at most max lat
        wn_train.iloc[rowA, 3] >= spray_centers.iloc[rowB, 5] and #and longitude is at least min long
        wn_train.iloc[rowA, 3] <= spray_centers.iloc[rowB, 6]): #and longitude is at most max long
            
            has_sprayed_dict[f'{spray_centers.iloc[rowB, 0]}'].append(1) #is the date is at or after the spray, and location matches
        
        else:
            
            has_sprayed_dict[f'{spray_centers.iloc[rowB, 0]}'].append(0) #is the date is at or after the spray, and location matches
            
            
```
